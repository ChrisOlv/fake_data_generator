```python
import pandas as pd
import random
from faker import Faker
import numpy as np
```

### V1

```python
fake = Faker('fr_FR')

# Define the number of entries
num_sales = 10000
num_clients = 50
num_vegetables = 50

# Generate random dates for sales
date_range = pd.date_range(start='2024-01-01', end='2024-12-31')

# Generate sales data
sales = pd.DataFrame({
    'ID_Vente': range(1, num_sales + 1),
    'ID_Client': [random.randint(1, num_clients) for _ in range(num_sales)],
    'ID_Légume': [random.randint(1, num_vegetables) for _ in range(num_sales)],
    'Date_Vente': [random.choice(date_range) for _ in range(num_sales)],
    'Quantité': [random.randint(1, 100) for _ in range(num_sales)]
})

# Extended list of fruits and vegetables to ensure sufficient unique names
extended_fruits_vegetables = [
    "Pomme", "Banane", "Orange", "Tomate", "Carotte", "Brocoli", "Poire", "Raisin", "Fraise", "Courgette",
    "Citron", "Chou-fleur", "Laitue", "Épinard", "Poivron", "Mangue", "Pêche", "Aubergine", "Ananas", "Concombre",
    "Framboise", "Myrtille", "Kiwi", "Pastèque", "Melon", "Cerise", "Grenade", "Papaye", "Litchi", "Cassis",
    "Artichaut", "Asperge", "Betterave", "Chou", "Champignon", "Haricot", "Navet", "Oignon", "Poireau", "Radis"
]

# Ensure the list is large enough
while len(extended_fruits_vegetables) < num_vegetables:
    extended_fruits_vegetables *= 2
extended_fruits_vegetables = extended_fruits_vegetables[:num_vegetables]

# Generate vegetables data
vegetables = pd.DataFrame({
    'ID_Légume': range(1, num_vegetables + 1),
    'Nom_Légume': random.sample(extended_fruits_vegetables, num_vegetables)
})

# Generate random names and companies for clients
clients = pd.DataFrame({
    'ID_Client': range(1, num_clients + 1),
    'Nom': [fake.name() for _ in range(num_clients)],
    'Entreprise': [fake.company() for _ in range(num_clients)]
})

# Save updated data to CSV
clients.to_csv('clients_updated.csv', index=False)
vegetables.to_csv('vegetables_updated.csv', index=False)
sales.to_csv('sales_updated.csv', index=False)
```
